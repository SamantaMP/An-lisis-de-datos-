```python

# Calcular la media de 'resultado' por 'ciudad' y ordenar
ordenado = datos.groupby('ciudad')['resultado'].mean().sort_values(ascending=False).index

# Ajusta el tama√±o de la figura
plt.figure(figsize=(17, 6))

# Crear un gr√°fico de barras agrupadas y ordenadas
sns.barplot(data=datos, x='ciudad', y='resultado', order=ordenado)
plt.title('Resultados por ciudad y resultado de la prueba de Anemia')
plt.xlabel('Ciudad')
plt.ylabel('Resultado')
plt.show()
```

```python

# Dividir los datos en caracter√≠sticas (X) y la variable objetivo (y)
X = datos.drop('resultado', axis=1)  # Caracter√≠sticas
y = datos['resultado']  # Variable objetivo

# Dividir los datos en conjuntos de entrenamiento y prueba (80% para entrenamiento, 20% para prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Definir las columnas categ√≥ricas y num√©ricas
columnas_categoricas = ['ciudad']  # Especifica las columnas categ√≥ricas
columnas_numericas = [col for col in X.columns if col not in columnas_categoricas]

# Definir el transformador para aplicar One-Hot Encoding a las variables categ√≥ricas
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(sparse_output=False), columnas_categoricas),  # Aplica One-Hot Encoding a las columnas categ√≥ricas, El nombre "onehot" es un identificador para este transformador
        ('passthrough', 'passthrough', columnas_numericas)  # Deja las columnas num√©ricas sin modificar 'passthrough' indica que las columnas deben dejarse sin modificar
    ])

# Aplicar el preprocesamiento a los datos de entrenamiento y prueba
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)


print(X_test)


# Inicializar y entrenar el modelo de regresi√≥n lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X_train_processed, y_train)
```
